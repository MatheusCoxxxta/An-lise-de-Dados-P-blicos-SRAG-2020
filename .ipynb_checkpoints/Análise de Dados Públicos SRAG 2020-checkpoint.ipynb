```python
import pandas as pd
import matplotlib.pyplot as plt
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = r'C:\Users\thais\INFLUD-07-07-2020.csv'
df = pd.read_csv(filename, delimiter=';',
                 encoding='ISO-8859-1',
                error_bad_lines=False)
```

```python
df.shape
```

```python
df.info
```

```python
df.sample(10)
```

```python
df.groupby('CS_SEXO').size().sort_values().plot(kind='bar', color='purple', figsize=(7,7), fontsize=(13))
plt.title('Incidência de Sindrome Respiratória por sexo', fontsize=(13))
```

```python
df.groupby('CS_RACA').size().sort_values().plot(kind='barh', color='violet', figsize=(7,7), fontsize=(13))
plt.title('Número de casos de SRAG por raça', fontsize=(13))
```

```python
df.groupby('NU_IDADE_N').size().sort_values().tail(10).plot(kind='bar', color='pink', figsize=(10,5), fontsize=(13))
plt.title('Idades com maior incidência de casos de SRAG', fontsize=(13))
```
